Matplotlib


*   used for making plots



```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
# sin wave
plt.plot(x,y)
plt.show()
```

```python
# cos wave
plt.plot(x,z)
plt.show()
```

```python
# adding titles, x-axis and y-axis labels
plt.plot(x,y)
plt.title('sine wave')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# parabola
x = np.linspace(-10,10,100)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
# plot with red color and + sign
plt.plot(x,y,'r+')
plt.show()
```

```python
# plot with green color and . sign
plt.plot(x,y,'g.')
plt.show()
```

```python
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x),'r-')
plt.plot(x,np.cos(x),'g.')
plt.show()
```

```python
# bar graph
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
Language = ['Java','Python','C++','C']
students = [23,17,35,32]
ax.bar(Language,students)
plt.xlabel('Languages')
plt.ylabel('No. of students')
plt.show()
```

```python
# pie chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
Language = ['Java','Python','C++','C']
students = [23,17,35,32]
ax.pie(students,labels=Language,autopct='%1.1f%%')
plt.show()
```

```python
# Scatter plot
x = np.linspace(0,10,100)
y = np.sin(x)
z=np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='r')
ax.scatter(x,z,color='g')
plt.show()
```
